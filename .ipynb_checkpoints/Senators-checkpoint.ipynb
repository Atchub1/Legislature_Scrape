```python
import docx
# from docx import Document
import pandas as pd

# document = Document()

doc = docx.Document()

senator_df = pd.read_csv('senator_csv.csv')
senator_df
```

```python
# senator_df.iloc[0,1]
```

```python
doc.add_paragraph(senator_df.iloc[0,1])
```

```python
# doc.add_picture('zophie.png', width=docx.shared.Inches(1), height=docx.shared.Cm(4))
# senator_df.iloc[0,3]

doc.add_picture(senator_df.iloc[0,3], width=docx.shared.Inches(1),height=docx.shared.Cm(4))
```

```python
# adds a page break
doc.paragraphs[0].runs[0].add_break(docx.text.run.WD_BREAK.PAGE)
```

```python
# working with the 2nd page
doc.add_paragraph(senator_df.iloc[1,1])
# doc.add_picture(senator_df.iloc[1,3], width=docx.shared.Inches(1),height=docx.shared.Cm(4))
```
